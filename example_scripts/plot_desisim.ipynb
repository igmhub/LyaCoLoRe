# Read and plot transmissions with desisim format

```python
%pylab inline
```

```python
import numpy as np
import fitsio
import matplotlib.pyplot as plt
```

```python
#Open data file (test file with few quasars)
lyafile='../example_data/desisim_lya/desisim_lya_N10.fits'
#lyafile='test_desisim_lya_N76304.fits'
h = fitsio.FITS(lyafile)
nqso = len(h)-1
print('# quasars =',nqso)
```

```python
plt.xlabel('wavelength [A]')
plt.ylabel('flux')
for i in range(5):
    data=h[i+1].read()
    la = data['LAMBDA'][:]
    tr = data['FLUX'][:]
    plt.plot(la,tr)
```
